```python
import os
import pandas as pd
import shutil
```

```python
labels = {"good weld"  :         0,
"burn through"   :    1,
"contamination"   :    2,
"lack of fusion"  :    3,
"misalignment"    :    4,
"lack of penetration" :5}
```

```python
inv_labels = {v: k for k, v in labels.items()}
```

```python
import json
  
# Opening JSON file
f = open('train/train.json')
  
# returns JSON object as 
# a dictionary
data = json.load(f)
  
# Iterating through the json
# list
# for i in data['emp_details']:
#     print(i)
  
# Closing file
# f.close()
```

```python
root_dir = 'train'
```

```python
for label in labels:
    if((os.path.exists(os.path.join(root_dir,label)))==False):
        os.mkdir(os.path.join(root_dir,label))
```

```python
from tqdm import tqdm
name_label = {}
for i in os.listdir(root_dir)[:-1]:
    for file_name in tqdm(os.listdir(os.path.join(root_dir,i))):
        move_folder = inv_labels[data[i+'/'+file_name]]
        shutil.copy(os.path.join(root_dir,i,file_name),os.path.join(root_dir,move_folder,file_name))
```
